### some notes on the way:
- create data (here: 'random' vars)
- create traces by 'go' statement
- make list out of data
- pass the list to the iplot function which creates the graph

```python
N = 100
random_x = np.linspace(0, 1, N)
random_y0 = np.random.randn(N) + 5
random_y1 = np.random.randn(N)
random_y2 = np.random.randn(N) - 5

trace0 = go.Scatter(x=random_x, y=random_y0, mode="markers", name="markers")
trace1 = go.Scatter(x=random_x, y=random_y1, mode="lines+markers", name="lines+markser")
trace2 = go.Scatter(x=random_x, y=random_y2, mode="lines", name="lines")

data = [trace0, trace1, trace2]

iplot(data)
```

### creating simple bar chart
- title
- labels at x a y axes
- fontsize of the title and axes labels
- color of the bars

- "font" controls all the font, "titlefont" only corresponding titles


```python
# creatign bars with their color
data = [go.Bar(x=["sample_1", "sample_2", "sample_3"], 
               y=[20, 40, 60],
              marker=dict(color="lightblue"))]


# creating layout
layout = go.Layout(title="some bars", titlefont=dict(size=25), font=dict(size=15),
                   xaxis=dict(title="sampling points", titlefont=dict(size=20)), 
                   yaxis=dict(title="count", titlefont=dict(size=20)))

# fig is the new obj to pass to iplot
fig = go.Figure(data=data, layout=layout)

iplot(fig)
```
