```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)


import numpy as np
```

### some notes on the way:
- create data (here: 'random' vars)
- create traces by 'go' statement
- make list out of data
- pass the list to the iplot function which creates the graph

```python
N = 100
random_x = np.linspace(0, 1, N)
random_y0 = np.random.randn(N) + 5
random_y1 = np.random.randn(N)
random_y2 = np.random.randn(N) - 5

trace0 = go.Scatter(x=random_x, y=random_y0, mode="markers", name="markers")
trace1 = go.Scatter(x=random_x, y=random_y1, mode="lines+markers", name="lines+markser")
trace2 = go.Scatter(x=random_x, y=random_y2, mode="lines", name="lines")

data = [trace0, trace1, trace2]

iplot(data)
```

```python
labels = ['Oxygen','Hydrogen','Carbon_Dioxide','Nitrogen']
values = [4500,2500,1053,500]

trace = go.Pie(labels=labels, values=values)

iplot([trace])
```

### creating simple bar chart
- title
- labels at x a y axes
- fontsize of the title and axes labels
- color of the bars

- "font" controls all the font, "titlefont" only corresponding titles


```python
# creatign bars with their color
data = [go.Bar(x=["sample_1", "sample_2", "sample_3"], 
               y=[20, 40, 60],
              marker=dict(color="lightblue"))]


# creating layout
layout = go.Layout(title="some bars", titlefont=dict(size=25), font=dict(size=15),
                   xaxis=dict(title="sampling points", titlefont=dict(size=20)), 
                   yaxis=dict(title="count", titlefont=dict(size=20)))

# fig is the new obj to pass to iplot
fig = go.Figure(data=data, layout=layout)

iplot(fig)
```

### Creating maps

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2011_us_ag_exports.csv')
```

```python
df.head(3)
```

```python
# converting values to str obj

for col in df.columns:
    df[col] = df[col].astype(str)
```

```python
scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]
```

```python
df["text"] = df["state"] + "<br>" + \
'Beef '+df['beef']+' Dairy '+df['dairy']+'<br>'+\
'Fruits '+df['total fruits']+' Veggies ' + df['total veggies']+'<br>'+\
'Wheat '+df['wheat']+' Corn '+df['corn']
```

```python
df.head(1)
```

```python

data = [ dict(type="choropleth",
             colorscale = scl,
             autocolorscale = False,
             locations = df["code"],
             z = df["total exports"].astype(float),
             locationmode = "USA-states",
             text = df["text"],
             # 
             marker = dict(line = dict(color="rgb(0,255,255)", width = 0)),
              colorbar = dict(title = "Millions USD"))]


layout = dict(title = "2011 US Agriculture Exports by State<br>(Hover for breakdown)",
             geo = dict(scope="usa",
                       projection=dict(type="albers usa"),
                       showlakes = True,
                       lakecolor = "rgb(255, 255, 255)"),)

fig = dict(data=data, layout=layout)
iplot(fig)
```
